```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
import pandas as pd
data = [['LR-BoW',0.826597], ['RF-BoW', 0.816541], ['DT-Bow', 0.781211],['KN-BOW',0.773678],['GB-BOW',0.701487],
        ['LR-TF-IDF',0.830070],['RF-TF-IDF',0.816916],['DT-TF-IDF',0.797185],['KN-TF-IDF',0.768583],['GB-TF-IDF',0.769883],
        ['LR-IndicNLP',0.76676], ['RF-IndicNLP', 0.874690], ['DT-IndicNLP', 0.798990],['KN-IndicNLP',0.846915],['GB-IndicNLP',0.766219]]
final_df = pd.DataFrame(data, columns = ['Model', 'Accuracy'])
final_df
```

```python
df=final_df
```

```python
font1 = {'size':30, 'weight': 'bold'}
font2 = {'size':25, 'weight': 'bold'}
font3 = {'size':20}
```

```python
X = ['LR (BoW)', 'RF (BoW)', 'DT (BoW)', 'KN (BoW)', 'GB (BoW)', 'LR (TF-IDF)', 'RF (TF-IDF)', 'DT (TF-IDF)', 'KN (TF-IDF)', 'GB (TF-IDF)', 'LR (IndicNLP)', 'RF (IndicNLP)', 'DT (IndicNLP)', 'KN (IndicNLP)', 'GB (IndicNLP)']
A = df['Accuracy']

X_axis = np.arange(len(X))
fig, ax =plt.subplots(figsize=(30, 18))

ax.bar(X_axis,A)
plt.xticks(X_axis, X, fontsize=25, rotation=90)
plt.yticks(fontsize=20)
plt.xlabel("Models", fontdict = font2)
plt.ylabel("Accuracy", fontdict = font2)
plt.title("Performance of various Models and Vectorization Techniques",fontdict = font1)
plt.rcParams['legend.fontsize'] = 25
plt.legend(loc = "lower right")
plt.show()
fig.savefig('Performance of various Models and Vectorization Techniques.pdf', format='pdf')
```

```python
data_bow = [['LR',0.826597], ['RF', 0.816541], ['DT', 0.781211],['KN',0.773678],['GB',0.701487]]
data_tfidf = [['LR',0.830070],['RF',0.816916],['DT',0.797185],['KN',0.76858],['GB',0.769883]]
data_indicNLP = [['LR-IndicNLP',0.936597], ['RF-IndicNLP', 0.874690], ['DT-IndicNLP', 0.798990],['KN-IndicNLP',0.846915],['GB-IndicNLP',0.766219]]
df_bow = pd.DataFrame(data_bow, columns = ['Model', 'Accuracy'])
df_tfidf = pd.DataFrame(data_tfidf, columns = ['Model','Accuracy'])
df_IndicNLP = pd.DataFrame(data_indicNLP, columns = ['Model','Accuracy'])
```

```python
X= ['LR','RF','DT','KN','GB']
A= df_bow['Accuracy']
B = df_tfidf['Accuracy']
c = df_IndicNLP['Accuracy']
models = df_bow['Model']
x = np.arange(len(models))
```
